```python
import urllib.request
d = urllib.request.urlopen("https://wikidocs.net")

data = d.read()
data = data.decode("utf-8")
print(data)
```

```python
import re
all_hangule = re.findall("[가-힣]+", data)
for i in all_hangule:
  print(i)
```

```python
import urllib.parse
import urllib.request

keyword = input('검색할 단어를 입력하세요 -> ')
keyword = urllib.parse.quote(keyword)

url = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query="+keyword
d = urllib.request.urlopen(url)

if d.status == 200:
  data = d.read().decode("utf-8")
  f = open("result.html", "w")
  f.write(data)
  f.close()
print(data)
```
